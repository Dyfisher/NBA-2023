```python
import pandas as pd
```

```python
data = pd.read_csv('2023_NBA.csv')
data.head()
```

```python
data.drop(columns=['RANK'], inplace=True)
```

```python
data.head()
```

**<span style="color:red; font-size:20px">Top 10 youngest AND oldest active NBA Players</span>**


```python
data.sort_values(by='AGE', ascending=False).head(10)
```

```python
data.sort_values(by='AGE', ascending=True).head(10)
```

 **<span style="color:red; font-size:20px">Top 10 Players with most assists per game</span>**

```python
data.sort_values(by='APG', ascending=False).head(10)
```

 **<span style="color:red; font-size:20px">Top 10 leading scorers</span>**

```python
data.sort_values(by='PPG', ascending=False).head(10)
```

 **<span style="color:red; font-size:20px">Top 10 leading players in points, rebounds, and assists (combined)</span>**

```python
data.sort_values(by='P+R+A', ascending=False).head(10)
```
